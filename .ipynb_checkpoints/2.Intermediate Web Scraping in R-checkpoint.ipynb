**Cascading Style Sheets**, or **CSS**, is a language for adding styles to HTML pages. We may have noticed that our simple HTML pages from the past few screens didn't have any styling (all of the paragraphs had black text and the same font size). Most Web pages use CSS to display a lot more than basic black text.

CSS uses **selectors** to add styles to the elements and classes of elements we specify. We can use selectors to add background colors, text colors, borders, padding, and many other style choices to the elements on HTML pages.

If we'd like to learn more about CSS, [MDN's guide](https://developer.mozilla.org/en-US/docs/Learn/CSS/First_steps) is a great place to start.

However, we can't wait to master the selectors to extract the data. Several solutions are available to solve this problem.

* Retrieve the expected data programmatically. This solution assumes that we understand the web page's structure.
* Use a tool to find the suitable selector.

In practice, we often find ourselves combining both approaches  especially since most people won't give us the code for their pages. We have to try to understand web pages' structures by ourselves.

Typically, there is a selector for everything we might want to extract, but sometimes it can be quite complex. In these cases, we have to think about programming solutions.

Basically, two kinds of tools are useful for scraping web pages:

* A tool to display the web page HTML code and its structure, such as [Browser Developer Tools](https://developer.mozilla.org/en-US/docs/Learn/Common_questions/What_are_browser_developer_tools#the_inspector_dom_explorer_and_css_editor)
* A tool to test the selectors directly on the web page to scrape

Every modern web browser includes a powerful suite of developer tools. These tools can help you inspect currently loaded HTML and understand its structure.

* With [Chrome](https://developers.google.com/web/tools/chrome-devtools/dom), we can double-click on the page and select `inspect`.
* With [Safari](https://coolestguidesontheplanet.com/safari-web-developer-tools-show-dock-browser-window/), Go to `Preferences > Advanced `and then check `"Show Develop menu in menu bar"`.
* With [Mozilla Firefox](https://developer.mozilla.org/en-US/docs/Learn/Common_questions/What_are_browser_developer_tools#the_inspector_dom_explorer_and_css_editor), we can double-click on the page and select `inspect element`.

This tool helps understand the web pages' structure and identify the data we want to extract.

The second tool we'll use is [Selector Gadget](https://selectorgadget.com/). It generates and discovers selectors in web pages (refer to the video on the [tool's page](https://selectorgadget.com/) to see how it works in practice.) If we're interested in installing it, the procedure is on [the page](https://selectorgadget.com/).

Let's continue to use the [Billboard Hot 100 page](https://www.billboard.com/charts/hot-100). We want to extract the number of weeks the top ten songs were on the chart at [the beginning of 2020](https://www.billboard.com/charts/hot-100/2020-01-04).

We can enter this selector in Selector Gadget to see which elements this selector will take into consideration.

```python
Image("selectorgadget.png")
```

```python
Image("selectorgadget_2.png")
```

With this selector in mind, after inspecting the code again, we realize that this `.chart-element__information .color--primary` selector works.

In practice, sometimes inspecting the code is enough; therefore, there's no need to use Selector Gadget. However, we recommend to use it often because a CSS selector can represent more or fewer elements (data) than we want to extract.

Let's use the selector we found to extract the titles of the top ten songs. Notice we will no longer need Selector Gadget to resolve the exercise

**Task**



**Answer**

`content_hot_100 <- read_html("https://www.billboard.com/charts/hot-100/2020-01-04")`

`hot_100_top_10_songs <- content_hot_100 %>% 
    html_nodes(".chart-element__information .color--primary") %>%
    html_text() %>%
    head(10)`

`hot_100_top_10_songs`
