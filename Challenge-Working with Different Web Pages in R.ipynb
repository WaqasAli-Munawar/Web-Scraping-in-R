**Web page description**: [Billboard Hot 100](https://www.billboard.com/charts/hot-100) displays the current Hot 100 songs and other information. Suppose we want to build an image dataset for a machine learning project. In this case, we need to find (many) images. We will extract them from websites.

**Our goal**: we want to extract the image links from the Hot 100 recorded on January 4, 2020.

The expected result is the links behind the following images:

```python
Image("hot100_images.png")
```

**Task**

We provided a regular expression(`url_pattern`) that we can use to extract URLs from characters.

1. From the image tag, extract `**style**` **attribute** values from [this web page](https://www.billboard.com/charts/hot-100/2020-01-04).

    * To avoid external servers instability issues, use this link: "http://dataquestio.github.io/web-scraping-pages/The%20Hot%20100%20Chart%20_%20Billboard.html", which is a copy of the study page on our servers.
    
2. Select only the first five **`style` attribute** values.

3. From each style attribute value, extract the image URL.


**Answer**

`url_pattern <- "(?i)http[s]?://(?:[a-z]|[0-9]|[$-_@.&+]|[!*\\(\\),]|(?:%[0-9a-f][0-9a-f]))+\\.jpg"`


`hot_100_styledata <- scraper(url = "http://dataquestio.github.io/web-scraping-pages/The%20Hot%20100%20Chart%20_%20Billboard.html", 
                             selector = ".chart-element__image", 
                             output = "style")`

`hot_100_styledata_top5 <- head(hot_100_styledata, 5)`

`hot_5_img_url <- stringr::str_extract(hot_100_styledata_top5, url_pattern)`
